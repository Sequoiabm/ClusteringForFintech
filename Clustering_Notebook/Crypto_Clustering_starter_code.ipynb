```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data)

market_data_scaled[0:3]
```

```python
# Create a DataFrame with the scaled data
market_df_scaled = pd.DataFrame(
    market_data_scaled, 
    columns= df_market_data.columns
    )

market_df_scaled.head()

# Copy the crypto names from the original data
market_df_scaled["coin_id"] = df_market_data.index

# Set the coinid column as index
market_df_scaled = market_df_scaled.set_index("coin_id")

# Display sample data
market_df_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
k
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k

for i in k:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=i, random_state=1)
    # 2. Fit the model to the data using `df_market_data_scaled`
    model.fit(market_df_scaled)
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_)

inertia_df = pd.DataFrame(inertia, columns =["Inertia Value: "])
inertia_df

```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_dict = {
    "k": k,
    "inertia": inertia
}


# Create a DataFrame with the data to plot the Elbow curve

elbow_data_df = pd.DataFrame(elbow_dict)
elbow_data_df
```

inertia in KMeans is like the score of how well the model is doing. It's like a game where the model tries to keep the points in each group as close together as possible. Lower score, better performance! 

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot = elbow_data_df.hvplot.line( x="k", 
                                        y="inertia", 
                                        title="Crypto Inertia vs. K Value"
                                        )
elbow_plot
```

**Question:** What is the best value for `k`?

**Answer:** The best value for 'k' is 4, this is where the inertia is 79.022435. at this In this case, the elbow point seems to be around k=4 because the inertia decreases significantly up to that point, but the reduction becomes less pronounced as you increase k further.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4)
```

```python
# Fit the K-Means model using the scaled data
model.fit(market_df_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
market_clusters = model.predict(market_df_scaled)

# Print the resulting array of cluster values.
market_clusters
```

```python
# Create a copy of the DataFrame
market_predictions_df = market_df_scaled.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
market_predictions_df["Predicted_Clusters"] = market_clusters

# Display sample data
market_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
clusters_plot = market_predictions_df.hvplot.scatter(x="price_change_percentage_24h", 
                                     y="price_change_percentage_7d",
                                     by="Predicted_Clusters",
                                     hover_cols=["coin_id"],
                                     marker=["hex", "square", "cross", "inverted_triangle"],
                                     title="Crypto Clusters"
                                     )
clusters_plot
```

(Note) above^
This code is making a cool scatter plot for cryptocurrencies. Imagine each point as a crypto, with its 24-hour price change on one axis, and its 7-day change on another. The plot groups them into clusters, like different teams, based on KMeans. When you hover over a point, it shows the crypto's name. It's like a map to see how cryptos behave over time, grouped by their performance.

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.

model = PCA(n_components=3)
model

```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
cryptodata = model.fit_transform(market_df_scaled)

# View the first five rows of the DataFrame. 
cryptodata[0:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
model.explained_variance_ratio_

```

```python
# Create a new DataFrame with the PCA data.
crypto_pca_df = pd.DataFrame(cryptodata, columns=["PC1", "PC2", "PC3"])


# Copy the crypto names from the original data
crypto_pca_df["coin_id"] = df_market_data.index

# Set the coinid column as index
crypto_pca_df = crypto_pca_df.set_index("coin_id")

# Display sample data
crypto_pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
k
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(crypto_pca_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_dict = {
    "k": k,
    "inertia": inertia
}
# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_dict)
elbow_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_plot_pca = elbow_df.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve using PCA",
    xticks=k
)
elbow_plot_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for 'k' would be 4 (4 clusters) when using the PCA data. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:**  No, both values are identical. 

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4)
model
```

```python
# Fit the K-Means model using the PCA data
model.fit(crypto_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
crypto_clusters = model.predict(crypto_pca_df)
# Print the resulting array of cluster values.
crypto_clusters
```

```python
# Create a copy of the DataFrame with the PCA data
pca_df = crypto_pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_df["predicted_clusters"] = crypto_clusters

# Display sample data
pca_df
# Copy the crypto names from the original data
crypto_names2 = market_predictions_df.index

pca_df = pca_df.set_index(crypto_names2)

# Set the coinid column as index
pca_df


```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

pca_cluster_plot = pca_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="predicted_clusters",
    hover_cols=["coin_id"],
    marker=["hex", "square", "cross", "inverted_triangle"],
    title="Crypto Clusters (Round 2)"
    
    
)
pca_cluster_plot
```
